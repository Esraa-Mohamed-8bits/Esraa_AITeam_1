```python
import matplotlib.pyplot as plt
```

```python
class LinearRegressionGD:

  def __init__(self):

    self.B0 = None
    self.B1 = None
    self.SSE = float('inf')
    self.MSE = None

  def fit(self, X, y, learning_rate=0.01, epochs=1000, epsilon=1e-6):

    self.n = len(X)
    self.X = np.array(X)
    self.y = np.array(y)

    #Initialize coefficients
    self.B0 = 0
    self.B1 = 0


    #Gradient Descent
    for _ in range(epochs):
      y_hat = self.B0 + self.B1 * self.X
      error = self.y - y_hat

      #Calculate gradients
      dB0 = (-2 / self.n) * np.sum(error)
      dB1 = (-2 / self.n) * np.sum(error * self.X)

      #Update coefficients
      self.B0 -= learning_rate * dB0
      self.B1 -= learning_rate * dB1

      #New SSE after update
      new_sse = np.sum((y - (self.B0 + self.B1 * X)) ** 2)

      #Convergence check
      if abs(new_sse - self.SSE) < epsilon:
        self.SSE = new_sse
        break

      self.SSE = new_sse

    #Calculate the mean square error
    self.MSE = self.SSE / self.n

  def predict(self, X):

    if self.B0 is None or self.B1 is None:

      raise ValueError("The model has not been fitted yet")

    X = np.array(X)
    return self.B0 + self.B1 * X

  def plot(self, X, y):

    if self.B0 is None or self.B1 is None:

      raise ValueError("The model has not been fitted yet.")

    #Ensure X and y are numpy arrays
    X = np.array(X)
    y = np.array(y)

    #Predicted values
    y_pred = self.predict(X)

    #Plot
    plt.scatter(X, y, color='blue', label='Data points')
    plt.plot(X, y_pred, color='red', label='Regression line')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title(f'Simple Linear Regression with line of best fit')
    plt.legend()
    plt.show()
```

```python
model = LinearRegressionGD()
```

```python
import numpy as np

# Generate random data
x = np.random.rand(100, 1) * 10  # 100 samples between 0 and 10
y = 2 * x + 5 + np.random.randn(100, 1) * 2 # y = 2x + 5 with some noise
```

```python
model.fit(x, y, 0.001, 1000000)
```

```python
print(f'Mean squared error is: {round(model.MSE,3)}')
```

```python
model.plot(x, y)
```
